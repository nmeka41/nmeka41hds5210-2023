### 2. Data Diversity

Use Python to identify how many unique values each column in your data has.  Use Python to identify which column has the greatest number of distinct values (aka the most diversity)?  If your data has a unique identifier (e.g. a person or visit or record ID) then find the column with the next most number of unique values.

```python
# Finding the number of unique values in each column of the loaded data
unique_values = data.nunique()

# Identifying the column with the greatest number of distinct values
max_unique_column = unique_values.idxmax()

unique_values, max_unique_column
print(max_unique_column)
```

### 4. Range of Values

Your data almost certainly has some numeric columns that can be summed or averaged.  Create a histogram that shows the distribution of values for this column.

```python
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv('/content/United_States_and_Peurto_Cancer_Data.csv')
if df['Count in Males'].dtype in ['int64', 'float64']:
    # Creating a histogram for the "Count in Males" column
    plt.figure(figsize=(10, 6))
    plt.hist(df['Count in Males'], bins=30, color='blue', alpha=0.7)
    plt.title('Distribution of Lung and Bronchus Cancer Cases in Males')
    plt.xlabel('Count in Males')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
else:
    print("The selected column is not numeric and cannot be used for a histogram.")
```
